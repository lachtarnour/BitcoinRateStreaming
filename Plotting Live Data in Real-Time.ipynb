```python
import requests
import time
import matplotlib.pyplot as plt
from datetime import datetime

prices = []
times = []

url = "https://binance43.p.rapidapi.com/avgPrice"

querystring = {"symbol":"ETHBTC"}

headers = {
"content-type": "application/octet-stream",
"X-RapidAPI-Key": "adf850cd33msh466f42a04b53560p1832fcjsn463f5a277945",
"X-RapidAPI-Host": "binance43.p.rapidapi.com"
}

plt.ion() # activate interactive mode
fig = plt.figure()

while True:
    response = requests.get(url, headers=headers, params=querystring)
    data = response.json()
    # add current price and time to the lists
    prices.append(float(data['price']))
    times.append(datetime.fromtimestamp(time.time()))

    # plot the data so far
    plt.clf() # clear the previous figure
    plt.plot(times, prices)

    # adjust x-axis to show the last 30 seconds
    plt.xlim(datetime.fromtimestamp(time.time()-600), datetime.fromtimestamp(time.time()))
    plt.xticks(rotation=90)

    # update the plot
    fig.canvas.draw()
    plt.pause(10)

plt.show() # show the final figure
```
